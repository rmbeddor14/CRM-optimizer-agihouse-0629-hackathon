### what is this

#### persist to data store
- turn data into json unstructured format
- embed with "all-MiniLM-L6-v2"
- upsert to datastax

#### query from data store
- embed query with "all-MiniLM-L6-v2"
- use datastax api to cosine similarity search
- return top result (returns all cell data)

````python
##import libs

import os  
from uuid import uuid4

import pandas as pd  
from sentence_transformers import SentenceTransformer  
from transformers import GPT2LMHeadModel, GPT2Tokenizer 

````

````python

# Load CSV
df = pd.read_csv('masked_data.csv')

# Transform to unstructured data, including additional fields
unstructured_data = [f"Data: {row.to_dict()}" for index, row in df.iterrows()]

# Generate embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(unstructured_data)
````

````python
# Prepare documents for insertion, including additional metadata (last finished milestone and borr base income) 
documents = [
    {
        "_id": str(uuid4()),
        "text": data,
        "$vector": vectors,
        "last_finished_milestone": row["Last Finished Milestone"],
        "borr_base_income": row["Borr Base Income"],
    }
    for data, vectors, (_, row) in zip(unstructured_data, embeddings, df.iterrows())
]
````

````python
## check the num of dimensions
## this is so that we can create the collection on datastax (requires you to define dimensionality) 
## the result of this is 384 (the dimensionality of all-MiniLM-L6-v2) 
## but if you replace all-MiniLM-L6-v2, then you can run this to confirm the dimensionality of your embedding model 
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
dimensionality = model.get_sentence_embedding_dimension()
print(f"The dimensionality of the embeddings is: {dimensionality}")
````

### Connect to Datastax 

- to do this part you'll need to sign up for a datastax (astradb) serverless account
- [astra db quickstart](https://docs.datastax.com/en/astra-db-serverless/get-started/quickstart.html)
- from the datastax ui, you can find:
- your api key (format: AstraCA:xyz)
- and your api endpoint url (format: https://xyz-us-east-2.apps.astra.datastax.com)
- add these into environment variables on your shell


#### how to add an env variable for your api key and api endpoint

1. using your bash shell

```
export DATASTAX_API_KEY="your api key"
export DATASTAX_API_ENDPOINT="your api endpoint"
```

2. using your jupyter notebook
```
import os

# Set an environment variable
os.environ['DATASTAX_API_KEY'] = {your api key}
os.environ['DATASTAX_API_ENDPOINT'] = {your api endpoint}
```
  

````python
### connect to datastax
### see above for info on how to 

from astrapy import DataAPIClient

# Initialize the client
client = DataAPIClient(os.environ.get('DATASTAX_API_KEY'))
db = client.get_database_by_api_endpoint(
  os.environ.get('DATASTAX_API_ENDPOINT')
)

print(f"Connected to Astra DB: {db.list_collection_names()}")
````

````python
## create a collection
## only do it once
collection = db.create_collection("masked_data_with_metadata", dimension=384)
````

````python
## get collection (do this if collection already created)

collection = db.get_collection("masked_data_with_metadata")
````

````python
### try to bulk upload

response = collection.insert_many(
documents
)
````

#### check querying


````python
## query part
query = "who has 25k gift"
````

````python
# Generate embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')
query_vector = model.encode(query).tolist()
````

````python
result = collection.find_one({}, sort={"$vector": query_vector})
print(result)
````

````python
### another query
query = "who is waiting for borrower to sign doc"
````

````python
# Generate embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')
query_vector = model.encode(query).tolist()
````

````python
result = collection.find_one({}, sort={"$vector": query_vector})
print(result)
````

````python
## top 5 results

results = collection.find({},
    sort={"$vector": query_vector}, limit=5
)

for doc in results:
  print(doc)
````

````python
### try a query with a metadata filter 

## query part
query = "who is waiting for a gift with an income under {insert income}"
income_threshold = 2000

# Generate embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')
query_vector = model.encode(query).tolist()

# get top 3 
results = collection.find(
    {"$and": [
        {"borr_base_income": {"$lt": income_threshold}}
    ]},
    sort={"$vector": query_vector}
)

for doc in results:
  print(doc)
````
